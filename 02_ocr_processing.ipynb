```python
import pytesseract
import cv2
import os
```

```python

pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python

video_path = "sample_video/videoplayback.mp4"
frame_output_dir = "slides_ocr/frames"
os.makedirs(frame_output_dir, exist_ok=True)
```

```python

video = cv2.VideoCapture(video_path)
fps = int(video.get(cv2.CAP_PROP_FPS))
interval_sec = 5  # Save one frame every 5 seconds
frame_interval = fps * interval_sec
```

```python

frame_count = 0
saved = 0
```

```python

while True:
    success, frame = video.read()
    if not success:
        break
    if frame_count % frame_interval == 0:
        frame_path = os.path.join(frame_output_dir, f"frame_{saved:04d}.jpg")
        cv2.imwrite(frame_path, frame)
        saved += 1
    frame_count += 1
```

```python

video.release()
print(f"✅ Saved {saved} frames to {frame_output_dir}")
```

```python

ocr_output_path = "slides_ocr/extracted_slide_text.txt"
all_slide_text = ""
```

```python

for file in sorted(os.listdir(frame_output_dir)):
    if file.endswith(".jpg"):
        frame_path = os.path.join(frame_output_dir, file)

        img = cv2.imread(frame_path)
        
        text = pytesseract.image_to_string(img)

        all_slide_text += text + "\n"
```

```python

with open(ocr_output_path, "w", encoding="utf-8") as f:
    f.write(all_slide_text)

print(f"✅ OCR output saved at: {ocr_output_path}")
```
